```python
import pandas as pd
# Creating DataFrame using list of lists
data = [
[1, 'sai', 25],
[2, 'ram', 30],
[3, 'Charan', 28]
]
df = pd.DataFrame(data, columns=['ID', 'Name', 'Age'])
print(df)
```

```python
#creating DataFrame using Dictionary
import pandas as pd
data={
    'Name':['sai','ram','charan'],
    'Age':['22','23','24'],
    'city':['hyd','hyd','hyd']
}
df=pd.DataFrame(data)
print(df)
```
