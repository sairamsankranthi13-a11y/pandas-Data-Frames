```python
#using relational operator
#filtering single cloumn
import pandas as pd
df=pd.read_csv("sales4.csv")
con=df['Product_Cost'] < 59000
df1=df[con]
print(df1)
```

```python
#filtering multiple column
import pandas as pd
df=pd.read_csv("sales4.csv")
con=df['Product_Cost'] > 59000
con1=df['Product_Cost'] > 69000
df1=df[con&con1]

print(df1)
```

```python
#filtering data frame by using idex location
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.iloc[:10,]
print(df1)
```

```python
import pandas as pd
df=pd.read_csv("sales4.csv")
rows=df.index[0:]
cols=["Product_Name","Customer_Id"]
df1=df.loc[rows,cols]
print(df1)
```

```python
#selcting multiple values of a column
import pandas as pd
df=pd.read_csv("sales4.csv")
a=df.Product_Name=='LG Washing Machine'
b=df.Customer_Id==1
c=a|b
df1=df[c]
print(df1)
```

```python
#isin()method
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df[df.Product_Name.isin(['LG Washing Machine'])]
print(df1)
```

```python
#unique()function
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.Product_Name.unique
print(df1)
```
