```python
import pandas as pd
a=pd.read_csv("Customer.csv")
b=pd.read_csv("Order.csv")
df=pd.DataFrame(a)
df1=pd.DataFrame(b)

# Strip whitespace from column names to prevent KeyError due to hidden spaces
df.columns = df.columns.str.strip()
df1.columns = df1.columns.str.strip()

inn_join=pd.merge(df,df1,on='CustomerID',how='inner')
print(inn_join)
```

```python
#leftjoin
import pandas as pd

a=pd.read_csv("Customer.csv")
b=pd.read_csv("Order.csv")
df=pd.DataFrame(a)
df1=pd.DataFrame(b)

# Strip whitespace from column names to prevent KeyError due to hidden spaces
df.columns = df.columns.str.strip()
df1.columns = df1.columns.str.strip()

left_join=pd.merge(df,df1,on='CustomerID',how='left')
print(left_join)
```

```python
#right
import pandas as pd

a=pd.read_csv("Customer.csv")
b=pd.read_csv("Order.csv")
df=pd.DataFrame(a)
df1=pd.DataFrame(b)

# Strip whitespace from column names to prevent KeyError due to hidden spaces
df.columns = df.columns.str.strip()
df1.columns = df1.columns.str.strip()

right_join=pd.merge(df,df1,on='CustomerID',how='right')
print(right_join)
```

```python
#outerjoin
import pandas as pd

a=pd.read_csv("Customer.csv")
b=pd.read_csv("Order.csv")
df=pd.DataFrame(a)
df1=pd.DataFrame(b)

# Strip whitespace from column names to prevent KeyError due to hidden spaces
df.columns = df.columns.str.strip()
df1.columns = df1.columns.str.strip()

outer_join=pd.merge(df,df1,on='CustomerID',how='outer')
print(outer_join)
```

```python
#concat(p)
import pandas as pd
a=[[1,2],[4,5]]
b=[[7,8],[10,11]]
c=['sai','ram']
df=pd.DataFrame(a,columns=c)
df1=pd.DataFrame(b,columns=c)
print(df)
print(df1)
```

```python
#concatenating one dataframe to another dataframe
import pandas as pd
a=[[1,2],[4,5]]
b=[[7,8],[10,11]]
c=['sai','ram']
df=pd.DataFrame(a,columns=c)
df1=pd.DataFrame(b,columns=c)
result=[df,df1]
df2=pd.concat(result)
print(df)
print()
print(df1)
print()
print(df2)
```

```python
import pandas as pd
a=[[1,2],[4,5]]
b=[[7,8],[10,11]]
c=['sai','ram']
df=pd.DataFrame(a,columns=c)
df1=pd.DataFrame(b,columns=c)
result=[df,df1]
df2=pd.concat(result,ignore_index=True)
print(df)
print()
print(df1)
print()
print(df2)
```

```python
import pandas as pd
a=[[1,2],[4,5]]
b=[[7,8],[10,11],[12,14]]
c=['sai','ram']
c1=['x','y']
df=pd.DataFrame(a,columns=c)
df1=pd.DataFrame(b,columns=c1)
result=[df,df1]
df2=pd.concat(result,axis=1)
print(df)
print()
print(df1)
print(df2)
```
