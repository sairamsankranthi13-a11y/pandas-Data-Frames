```python
#date  format parameter
import pandas as pd
d={
    'Product':['samsung','iphone','mi'],
    'Status':['delivered','delivered','in-transit'],
    'Cost':[100000,50000,15000],
    'Date':['20190902','20190903','20190904']
}
df=pd.DataFrame(d)
print(df.head())
print()
print(df.dtypes)
```

```python
#converting date with specific format
import pandas as pd
d={
    'Product':['samsung','iphone','mi'],
    'Status':['delivered','delivered','in-transit'],
    'Cost':[100000,50000,15000],
    'Date':['20190902','20190903','20190904']
}
df=pd.DataFrame(d)
df['Date']=pd.to_datetime(df['Date'],format='%Y%m%d')
print(df.head())
```

```python
#NaT values
import pandas as  pd
d={
    'Product':['samsung','iphone','mi'],
    'Status':['delivered','delivered','in-transit'],
    'Cost':[100000,50000,15000],
    'Date':['02-Aug-2019','Here date is misssing','25-Aug-219']
}
df=pd.DataFrame(d)
print(df.head())
print()
print(df.dtypes)
```

```python
#NaT values
import pandas as  pd
d={
    'Product':['samsung','iphone','mi'],
    'Status':['delivered','delivered','in-transit'],
    'Cost':[100000,50000,15000],
    'Date':['02-Aug-2019','Here date is misssing','25-Aug-219']
}
df=pd.DataFrame(d)
df['Date']=pd.to_datetime(df['Date'],format='%d-%b-%Y',errors='coerce')
print(df.head())
print()
print(df.dtypes)
```
