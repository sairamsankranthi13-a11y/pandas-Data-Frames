```python
#filling nan values with mean value
import pandas as pd
import numpy as np
data=[["sai",23,40000],
["ram",24,50000],
["prasad",25,55000],
["mahi",np.nan,90000],
["nandhu",np.nan,566660]]
df=pd.DataFrame(data,columns=["name","age","salary"])
print(df)
m=df['age'].mean()
df['age']=df['age'].fillna(m)
print()
print(df)
```
